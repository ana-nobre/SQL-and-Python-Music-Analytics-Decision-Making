#### The `artist.getInfo` call (Last.fm API) has intermittently returned status.code 200 with an empty body.

#### In these cases, the response headers show `Content-Length: 0`. To guard against this, such responses are filtered out with:

`response.headers.get('Content-Length') == '0'`




```python
import requests

url = 'https://ws.audioscrobbler.com/2.0/'
params = {
    'method': 'artist.getinfo',
    'artist' : 'Cher',                          
    'api_key': 'a7549be8afdcadd1a198425e381d26cf',
    'format': 'json'
    }

response = requests.get(url, params=params)
print(response.headers)
print(response.status_code)
responseJson = response.json()
responseJson
```
