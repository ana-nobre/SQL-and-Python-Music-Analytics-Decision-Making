```python
import csv
import mysql.connector

generos_csv = {
    1: 'list_artist_classical.csv',
    2: 'list_artist_jazz.csv',
    3: 'list_artist_pop.csv',
    4: 'list_artist_rock.csv'
}

cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1', database='Musicproject')
mycursor = cnx.cursor()

for id_genero, archivo_csv in generos_csv.items():
    print(f"Cargando artistas desde {archivo_csv} (Género {id_genero})...")

    with open(archivo_csv, newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        for row in reader:
            nombre_artista = row[0].strip()

            # Buscar datos en tabla_tracks
            mycursor.execute("""
                SELECT DISTINCT album, fecha, tipo, track
                FROM tabla_tracks
                WHERE nombre_artista = %s
                LIMIT 1
            """, (nombre_artista,))
            resultado = mycursor.fetchone()

            if resultado:
                album, fecha_raw, tipo, track = resultado

                # Corregir formato de fecha si es sólo año
                try:
                    if fecha_raw and len(str(fecha_raw)) == 4:
                        fecha = f"{fecha_raw}-01-01"
                    else:
                        fecha = fecha_raw
                except:
                    fecha = None

                insert_query = """
                    INSERT INTO tabla_artista (
                        id_genero, nombre_artista, album, fecha, tipo, track
                    ) VALUES (%s, %s, %s, %s, %s, %s)
                """
                valores = (id_genero, nombre_artista, album, fecha, tipo, track)

                try:
                    mycursor.execute(insert_query, valores)
                except mysql.connector.Error as err:
                    print(f" Error insertando {nombre_artista}: {err}")
            else:
                print(f"No se encontraron datos en tabla_tracks para: {nombre_artista}")

cnx.commit()
print("Todos los artistas fueron insertados.")

mycursor.close()
cnx.close()
```
